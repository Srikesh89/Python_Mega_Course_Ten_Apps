##### Converting location to coordinate

```python
import geopy
dir(geopy)
```

```python
from geopy.geocoders import ArcGIS
gis = ArcGIS()
g = gis.geocode("Speedway, Austin, TX")
g
```

```python
from geopy.geocoders import ArcGIS
gis = ArcGIS()
dataframe = pandas.read_csv("./supermarkets/supermarkets-commas.txt")
# dataframe
dataframe["Address"] = dataframe["Address"]+","+dataframe["City"]+","+dataframe["State"]+","+dataframe["Country"]
# dataframe
dataframe["Coordinates"] = dataframe["Address"].apply(gis.geocode)
dataframe["Latitude"] = dataframe["Coordinates"].apply(lambda x: x.latitude)
dataframe["Longitude"] = dataframe["Coordinates"].apply(lambda x: x.longitude)
dataframe
```
